```python
def levelgrab(pair):
    
    levslist = []
    
    for x in pair:
        Srates = mt5.copy_rates_from_pos(x, mt5.TIMEFRAME_H1, 0, 1000)
        S = Srates
        SCM = np.average(abs(S['close'] - S['open']))
        y = S['close']
        LVL = np.concatenate((y[argrelextrema(y, np.greater)[0]],y[argrelextrema(y, np.less)[0]]))
        B = list(combinations(LVL, 2))
        Ba = np.array([x for x, y in B])
        B = np.array([abs(x - y) for x, y in B])
        Bb = np.array([np.where(x <= SCM, 1, 0) for x in B])
        T = [(Ba[i], Bb[i]) for i in range(0, len(Ba))] 
        Toutput = defaultdict(int) 
        for k, v in T: 
            Toutput[k] += v 
        table = list(Toutput.items())
        tablevs = [x for x, y in table]
        tabpings = [y for x, y in table]
        tabmax = argrelextrema(np.array(tabpings), np.greater)
        tabmax2 = list(tabmax[0])
        levs = np.array(tablevs)[tabmax2]
        levs = list(levs)
        levs.sort(reverse=True)
        levs = list(levs)
        
        mindist = min([abs(j-i) for i, j in zip(levs[:-1], levs[1:])])
        
        while mindist < SCM:
            i = 0
            h = -1
            j = 1
            for c in range(0, len(levs)):
                if j == len(levs):
                    j = -1
                if abs(levs[i]-levs[h]) < SCM:
                    levs[i] = 0
                elif abs(levs[i]-levs[j]) < SCM:
                    levs[i] = levs[i] - abs(levs[i]-levs[j])/2
                i += 1
                h += 1
                j += 1
            levs = [v for v in levs if v != 0]
            mindist = min([abs(a-b) for a, b in zip(levs[:-1], levs[1:])])
            
        levs = tuple(levs)

        levslist.append(levs)

    return levslist
```

```python
def KHAN(pair, levslist):
    
    ia = 0
    
    for p in pair:
    
        G = MetaLink_pos(p,"M5",60)


        RED = []
        BLUE = []
        CLOUD = []
        CLOUDA = []
        CLOUDB = []
        PCLOUD = []
        REDBLUE = []
        PRED = []
        PBLUE = []
        SRED = []
        SBLUE = []
        SCORE_RED_LINE = []
        SCORE_PRICE_RED = []
        SCORE_BLUE_LINE = []
        SCORE_PRICE_BLUE = []
        SCORE_RED_BLUE = []
        SCORE_CLOUD = []
        SCORE_PRICE_CLOUD = []
        SCORE = []

        i = 0
        for x in range(0,len(G['close'])):
            if i < 9:
                RLhigh = 1
                RLlow = 1
            else:
                RL = G[i:i+9]
                RLhigh = np.max(RL['high'])
                RLlow = np.min(RL['low'])
            redline = (RLhigh + RLlow)/2
            RED.append(redline)

            if i < 26:
                BLhigh = 1
                BLlow = 1
            else:
                BL = G[i:i+26]
                BLhigh = np.max(BL['high'])
                BLlow = np.min(BL['low'])
            blueline = (BLhigh + BLlow)/2
            BLUE.append(blueline)

            if i < 52:
                CLhigh = 1
                CLlow = 1
            else:
                CL = G[i:i+52]
                CLhigh = np.max(CL['high'])
                CLlow = np.min(CL['low'])
            clouda = (redline + blueline)/2
            cloudb = (CLhigh + CLlow)/2
            CLOUDA.append(clouda)
            CLOUDB.append(cloudb)

            if clouda - cloudb > 0:
                cloud = 1
            elif clouda - cloudb < 0:
                cloud = -1
            else:
                cloud = 0
            CLOUD.append(cloud)

            if G['close'][i] - clouda > 0:
                pclouda = 1
            elif G['close'][i] - clouda < 0:
                pclouda = -1
            else:
                pclouda = 0
            if G['close'][i] - cloudb > 0:
                pcloudb = 1
            elif G['close'][i] - cloudb < 0:
                pcloudb = -1
            else:
                pcloudb = 0
            if pclouda == pcloudb:
                pcloud = pclouda
            else:
                pcloud = 0
            PCLOUD.append(pcloud)

            if redline > blueline:
                redblue = 1
            elif redline < blueline:
                redblue = -1
            else:
                redblue = 0
            REDBLUE.append(redblue)

            if G['close'][i] - redline > 0:
                pred = 1
            elif G['close'][i] - redline < 0:
                pred = -1
            else:
                pred = 0
            PRED.append(pred)

            if G['close'][i] - blueline > 0:
                pblue = 1
            elif G['close'][i] - blueline < 0:
                pblue = -1
            else:
                pblue = 0
            PBLUE.append(pblue)

            if i == 0:
                red_slope = 0
                blue_slope = 0
            else:
                if RED[i] - RED[i-1] > 0:
                    red_slope = 1
                if RED[i] - RED[i-1] < 0:
                    red_slope = -1
                else:
                    red_slope = 0
                if BLUE[i] - BLUE[i-1] > 0:
                    blue_slope = 1
                if BLUE[i] - BLUE[i-1] < 0:
                    blue_slope = -1
                else:
                    blue_slope = 0
            SRED.append(red_slope)
            SBLUE.append(blue_slope)

            if i == 0:
                score_red_line = 0
                score_price_red = 0
                score_blue_line = 0
                score_price_blue = 0
                score_red_blue = 0
                score_cloud = 0
                score_pcloud = 0
            else:
                if SRED[i] == 1:
                    if SRED[i] != SRED[i-1]:
                        score_red_line = 3
                    else:
                        score_red_line = 0
                elif SRED[i] == -1:
                    if SRED[i] != SRED[i-1]:
                        score_red_line = -3
                    else:
                        score_red_line = 0
                if PRED[i] == 1:
                    if PRED[i] != PRED[i-1]:
                        score_price_red = 3
                    else:
                        score_price_red = 0
                elif PRED[i] == -1:
                    if PRED[i] != PRED[i-1]:
                        score_price_red = -3
                    else:
                        score_price_red = 0
                if SBLUE[i] == 1:
                    if SBLUE[i] != SBLUE[i-1]:
                        score_blue_line = 8
                    else:
                        score_blue_line = 0
                elif SBLUE[i] == -1:
                    if SBLUE[i] != SBLUE[i-1]:
                        score_blue_line = -8
                    else:
                        score_blue_line = 0
                if PBLUE[i] == 1:
                    if PBLUE[i] != PBLUE[i-1]:
                        score_price_blue = 5
                    else:
                        score_price_blue = 0
                elif PBLUE[i] == -1:
                    if PBLUE[i] != PBLUE[i-1]:
                        score_price_blue = -5
                    else:
                        score_price_blue = 0
                if REDBLUE[i] == 1:
                    if REDBLUE[i] != REDBLUE[i-1]:
                        score_red_blue = 4
                    else:
                        score_red_blue = 0
                elif REDBLUE[i] == -1:
                    if REDBLUE[i] != REDBLUE[i-1]:
                        score_red_blue = -4
                    else:
                        score_red_blue = 0
                if CLOUD[i] == 1:
                    if CLOUD[i] != CLOUD[i-1]:
                        score_cloud = 5
                    else:
                        score_cloud = 0
                elif CLOUD[i] == -1:
                    if CLOUD[i] != CLOUD[i-1]:
                        score_cloud = -5
                    else:
                        score_cloud = 0
                if PCLOUD[i] == 1:
                    if PCLOUD[i] != PCLOUD[i-1]:
                        score_pcloud = 9
                    else:
                        score_pcloud = 0
                elif PCLOUD[i] == -1:
                    if PCLOUD[i] != PCLOUD[i-1]:
                        score_pcloud = -9
                    else:
                        score_pcloud = 0                    
            SCORE_RED_LINE.append(score_red_line)
            SCORE_PRICE_RED.append(score_price_red)
            SCORE_BLUE_LINE.append(score_blue_line)
            SCORE_PRICE_BLUE.append(score_price_blue)
            SCORE_RED_BLUE.append(score_red_blue)
            SCORE_CLOUD.append(score_cloud)
            SCORE_PRICE_CLOUD.append(score_pcloud)

            if i > 0:
                if SCORE_RED_LINE[i-1] != 0:
                    if SCORE_RED_LINE[i-1] == 1:
                        SCORE_RED_LINE[i] = 0
                    elif SCORE_RED_LINE[i-1] == -1:
                        SCORE_RED_LINE[i] = 0
                    else:
                        SCORE_RED_LINE[i] = SCORE_RED_LINE[i-1]/2
                if SCORE_PRICE_RED[i-1] != 0:
                    if SCORE_PRICE_RED[i-1] == 1:
                        SCORE_PRICE_RED[i] = 0
                    elif SCORE_PRICE_RED[i-1] == -1:
                        SCORE_PRICE_RED[i] = 0
                    else:
                        SCORE_PRICE_RED[i] = SCORE_PRICE_RED[i-1]/2
                if SCORE_BLUE_LINE[i-1] != 0:
                    if SCORE_BLUE_LINE[i-1] == 1:
                        SCORE_BLUE_LINE[i] = 0
                    elif SCORE_BLUE_LINE[i-1] == -1:
                        SCORE_BLUE_LINE[i] = 0
                    else:
                        SCORE_BLUE_LINE[i] = SCORE_BLUE_LINE[i-1]/2
                if SCORE_PRICE_BLUE[i-1] != 0:
                    if SCORE_PRICE_BLUE[i-1] == 1:
                        SCORE_PRICE_BLUE[i] = 0
                    elif SCORE_PRICE_BLUE[i-1] == -1:
                        SCORE_PRICE_BLUE[i] = 0
                    else:
                        SCORE_PRICE_BLUE[i] = SCORE_PRICE_BLUE[i-1]/2
                if SCORE_RED_BLUE[i-1] != 0:
                    if SCORE_RED_BLUE[i-1] == 1:
                        SCORE_RED_BLUE[i] = 0
                    elif SCORE_RED_BLUE[i-1] == -1:
                        SCORE_RED_BLUE[i] = 0
                    else:
                        SCORE_RED_BLUE[i] = SCORE_RED_BLUE[i-1]/2
                if SCORE_CLOUD[i-1] != 0:
                    if SCORE_CLOUD[i-1] == 1:
                        SCORE_CLOUD[i] = 0
                    elif SCORE_CLOUD[i-1] == -1:
                        SCORE_CLOUD[i] = 0
                    else:
                        SCORE_CLOUD[i] = SCORE_CLOUD[i-1]/2
                if SCORE_PRICE_CLOUD[i-1] != 0:
                    if SCORE_PRICE_CLOUD[i-1] == 1:
                        SCORE_PRICE_CLOUD[i] = 0
                    elif SCORE_PRICE_CLOUD[i-1] == -1:
                        SCORE_PRICE_CLOUD[i] = 0
                    else:
                        SCORE_PRICE_CLOUD[i] = SCORE_PRICE_CLOUD[i-1]/2

            score = SCORE_RED_LINE[i] + SCORE_PRICE_RED[i] + SCORE_BLUE_LINE[i] + SCORE_PRICE_BLUE[i] + SCORE_RED_BLUE[i] + SCORE_CLOUD[i] + SCORE_PRICE_CLOUD[i]
            SCORE.append(score)
            i +=1

        if score > 10:
            position = "LONG"
        elif score < -10:
            position = "SHORT"
        else:
            position = 0

        if position != 0:
            
            entry = G['close'][-1]
            levels = levslist[ia]
            lastone = abs(levels[0] - entry)
            q = 0
            for x in levels:
                if q == 0:
                    Q = q
                leveldiff = abs(x - entry)
                if leveldiff < lastone:
                    Q = q
                lastone = leveldiff
                q += 1
            CTESTLVL = levels[Q]

            cend = int(len(levels)-1)
            if position == 'LONG':
                clindex = levels.index(CTESTLVL)
                tpindex = int(clindex - 1)
                slindex = int(clindex + 1)
                if clindex == 0:
                    stoploss = levels[slindex]
                    sldist = entry - stoploss
                    takeprofit = entry + sldist
                elif clindex == cend:
                    takeprofit = levels[tpindex]
                    tpdist = takeprofit - entry
                    stoploss = entry - tpdist
                else:
                    takeprofit = levels[tpindex]
                    stoploss = levels[slindex]
            elif position == 'SHORT':
                clindex = levels.index(CTESTLVL)
                tpindex = int(clindex + 1)
                slindex = int(clindex - 1)
                if clindex == 0:
                    takeprofit = levels[tpindex]
                    tpdist = entry - takeprofit
                    stoploss = entry + tpdist
                elif clindex == cend:
                    stoploss = levels[slindex]
                    sldist = stoploss - entry
                    takeprofit = entry - sldist
                else:
                    takeprofit = levels[tpindex]
                    stoploss = levels[slindex]
            else:
                takeprofit = 0
                stoploss = 0

            shrink = 0.90
            if position == 'LONG':
                tpdist = takeprofit - entry
                takeprofit = entry + tpdist*shrink
            elif position == 'SHORT':
                tpdist = entry - takeprofit
                takeprofit = entry - tpdist*shrink

            if position == 'LONG':
                tpdist = takeprofit - entry
                sldist = entry - stoploss
                if sldist > tpdist:
                    stoploss = entry - tpdist
            elif position == 'SHORT':
                tpdist = entry - takeprofit
                sldist = stoploss - entry
                if sldist > tpdist:
                    stoploss = entry + tpdist

            if position == 'LONG':
                request, lotflag = positioner(p,"LONG",entry,takeprofit,stoploss)
                if lotflag == "GREEN":
                    result = mt5.order_send(request)
                    if result.retcode != mt5.TRADE_RETCODE_DONE:
                        print("order_send failed, retcode={}".format(result.retcode))
                        result_dict=result._asdict()
                        for field in result_dict.keys():
                            print("   {}={}".format(field,result_dict[field]))
                            if field=="request":
                                traderequest_dict=result_dict[field]._asdict()
                                for tradereq_filed in traderequest_dict:
                                    print("       traderequest: {}={}".format(tradereq_filed,traderequest_dict[tradereq_filed]))
                    else:
                        print("Trade Placed:   ",p,"   ","Long","   TP:  ",takeprofit,"   SL:  ",stoploss,)
            elif position == 'SHORT':
                request, lotflag = positioner(p,"SHORT",entry,takeprofit,stoploss)
                if lotflag == "GREEN":
                    result = mt5.order_send(request)
                    if result.retcode != mt5.TRADE_RETCODE_DONE:
                        print("order_send failed, retcode={}".format(result.retcode))
                        result_dict=result._asdict()
                        for field in result_dict.keys():
                            print("   {}={}".format(field,result_dict[field]))
                            if field=="request":
                                traderequest_dict=result_dict[field]._asdict()
                                for tradereq_filed in traderequest_dict:
                                    print("       traderequest: {}={}".format(tradereq_filed,traderequest_dict[tradereq_filed]))
                    else:
                        print("Trade Placed:   ",p,"   ","Short","   TP:  ",takeprofit,"   SL:  ",stoploss,)
            print(position,entry,takeprofit,stoploss,lotflag)
        ia += 1
```
